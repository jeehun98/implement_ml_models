```python
def standardization(X):
  """
  표준화 함수
  Arg:
    X : 표준화할 데이터
  Return : 
    표준화 후 데이터
  """
  mean = np.mean(X, axis=0)
  std = np.std(X, axis=0)

  return (X-mean)/std
```

```python
class standardization:
  mean = 0
  std = 0

  def __init__(self):
    self.mean = 0
    self.std = 0

  def standardization(self, X):
    """
    표준화 함수
    Arg:
      X : 표준화할 데이터
    Return : 
      표준화 후 데이터
    """
    self.mean = np.mean(X, axis=0)
    self.std = np.std(X, axis=0)

    return (X-self.mean)/self.std

  def transform(self, X):
    return (X-self.mean)/self.std
```

```python
std = standardization()

std_x = std.standardization(X)
std_x
```

```python
np.mean(std_x)
```

```python
np.var(std_x)
```
